```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
(X_train, y_train), _ = fashion_mnist.load_data()
```

```python
dress_types = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
plt.figure(figsize=(12,6))
```

```python
for i in range(10):
  index = np.where(y_train == i)[0][0]
  plt.subplot(2, 5, i+1)
  plt.imshow(X_train[index], cmap='BuPu')
  plt.title(dress_types[i])
  plt.axis('off')
```

```python
plt.suptitle("One sample for each dresses")
```

```python
plt.show()
```
