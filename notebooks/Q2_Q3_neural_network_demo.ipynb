# Question 2 & 3 — Feedforward NN + Backprop + All 6 Optimizers

```python
# ── Imports ────────────────────────────────────────────────────────────
import sys, os
sys.path.append('.')   # neural_network.py must be in same folder

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from neural_network import NeuralNetwork

# ── Load & preprocess ─────────────────────────────────────────────────
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
X_train_full = X_train_full.reshape(-1, 784) / 255.0
X_test       = X_test.reshape(-1, 784)       / 255.0

# 10% validation split
val_size = int(0.1 * len(X_train_full))
X_val, y_val     = X_train_full[-val_size:], y_train_full[-val_size:]
X_train, y_train = X_train_full[:-val_size], y_train_full[:-val_size]

print(f'Train: {X_train.shape}  Val: {X_val.shape}  Test: {X_test.shape}')
```

```python
# ── Q2 Demonstration — flexible architecture ──────────────────────────
print('=== Q2: Flexible Architecture Demo ===')
for arch in [[64], [128, 64], [256, 128, 64]]:
    model = NeuralNetwork(hidden_layers=arch)
    out   = model.forward(X_train[:5])
    print(f'  hidden_layers={arch} → output shape {out.shape} ✅')
```

```python
# ── Q3 Demonstration — all 6 optimizers ──────────────────────────────
print('=== Q3: All 6 Optimizers (3 epochs quick test) ===')
OPTIMIZERS = ['sgd', 'momentum', 'nesterov', 'rmsprop', 'adam', 'nadam']
results    = {}

for opt in OPTIMIZERS:
    m = NeuralNetwork(hidden_layers=[128, 64], activation='relu', weight_init='xavier')
    h = m.train(X_train, y_train, epochs=3, learning_rate=0.001,
                batch_size=128, optimizer=opt, verbose=False)
    acc = m.accuracy(X_val, y_val)
    results[opt] = acc
    print(f'  {opt:10s} | val_acc: {acc*100:.2f}%')

print('\n✅ Q2 & Q3 Complete!')
```

```python
# ── Train best model for further questions ────────────────────────────
print('Training best model (Adam, [128,64], relu, xavier, 20 epochs)...')
best_model = NeuralNetwork(
    hidden_layers=[128, 64],
    activation='relu',
    weight_init='xavier'
)
history = best_model.train(
    X_train, y_train,
    epochs=20, learning_rate=0.001,
    batch_size=128, optimizer='adam',
    weight_decay=0.0005, verbose=True
)

print(f'\nTrain Acc : {best_model.accuracy(X_train, y_train)*100:.2f}%')
print(f'Val   Acc : {best_model.accuracy(X_val,   y_val)*100:.2f}%')
print(f'Test  Acc : {best_model.accuracy(X_test,  y_test)*100:.2f}%')
```
